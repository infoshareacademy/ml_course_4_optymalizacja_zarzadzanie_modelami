# Symulacje


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Losowe dane
data = np.random.uniform(0,1000,size=500)
```

```python
data
```

```python
# Wartość średnia
np.mean(data)
```

## Losowanie bez zwracania

```python
# Losowanie bez zwracania
no_replacement  = np.random.choice(data,size=100, replace=False)
```

```python
# Wartość średnia
np.mean(no_replacement)
```

## Losowanie ze zwracaniem

```python
# Losowanie ze zwracaniem
replacament = np.random.choice(data, size=100, replace=True)
```

```python
# Wartość średnia
np.mean(replacament)
```

## Symulacja

```python
# Bez zwracania
n = 100
res_no = [] 
for i in range(n):
    no_replacement = np.random.choice(data,size=20,replace=False)
    mean_value = np.mean(no_replacement)
    res_no.append(mean_value)
```

```python
# Ze zwracaniem
n = 100
res = [] 
for i in range(n):
    replacement = np.random.choice(data,size=20,replace=True)
    mean_value = np.mean(replacement)
    res.append(mean_value)
```

```python
# Porównanie rozkładów
sns.kdeplot(res_no, label ='no raplacament')
sns.kdeplot(res,label = 'replacament')
plt.legend()
plt.show()
```

```python
# Describe ze zwracaniem
pd.Series(res).describe()
```

```python
# Describe bez zwracania
pd.Series(res_no).describe()
```

```python
# 95% przedział ufnosci
pd.Series(res_no).quantile([0.025,0.975])
```
